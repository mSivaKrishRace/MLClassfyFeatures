```python
from pydub.silence import detect_nonsilent
from pydub.silence import detect_silence

from pydub.effects import strip_silence
from pydub.silence import split_on_silence
```

```python
path_to_input_files = '/home/lenovo/All5_KannadaDataAug2018/5Class/'
save_to_folder = '/home/lenovo/All5_KannadaDataAug2018/Siva_Split_Knd_Data/'

max_classes = 5

#Dont Edit
max_classes += 1;
```

```python
for i in range(1,max_classes):
        folder_name =  path_to_input_files + 'D' + str(i) + '/'
        path_list = os.listdir(folder_name)
        j_num = 0
        for j in path_list:
                j_num += 1
                if '.wav' in j:
                    sound = AudioSegment.from_wav(folder_name + j)
                    silence_ranges = detect_silence(sound, min_silence_len = 200, silence_thresh = -36, seek_step = 5)
                    chunks = []
                    temp3 = 0
                   

                    for start_i, end_i in silence_ranges:
                        temp1 = start_i;
                        temp2 = end_i;
                        start_i = min(temp3, start_i)
                        end_i = temp1;
                        chunks.append(sound[start_i:end_i])
                        temp3 = temp2; 

                    for k, chunk in enumerate(chunks):
                        chunk.export(save_to_folder + 'D' + str(i) + "/{0}.wav".format(k), format="wav")
```
