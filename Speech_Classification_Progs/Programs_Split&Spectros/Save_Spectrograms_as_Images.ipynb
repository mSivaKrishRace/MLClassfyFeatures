```python
# Run this cell to freshly extract Mel_Spectrogram for all Audio Clips

j_num  = 0

for i in range(1,max_classes):
        folder_name =  path_to_files + 'D' + str(i) + '/' #reading Input Path
        saveAs_file =  save_folder_path  + 'D' + str(i) + '/'
        path_list = os.listdir(folder_name)
        j_num = 0
        for j in path_list:
                j_num += 1
                if '.wav' in j:
                    y_, sr = librosa.load(folder_name + j)
                    try:
                        pylab.axis('off') # no axis
                        pylab.axes([0., 0., 1., 1.], frameon=False, xticks=[], yticks=[]) # Remove the white edge
                                            #mel_f =librosa.feature.melspectrogram(y=y_, sr=sr, n_mels=128, fmax=8000)
                        mel_f =librosa.feature.melspectrogram(y=y_, sr=sr)
                    
                        librosa.display.specshow(librosa.power_to_db(mel_f, ref=np.max))
                    except:
                        print ('Problem at melspectrogram Extraction line 3')
                        continue 
                    try:
                        J_is = str(j)+ "\b\b\b\b"
                        print(J_is)
                        file_name = saveAs_file + str(J_is) + ".jpg"
                        #print(file_name)
                    except:
                        print ('Problem at file_name line 4')
                        continue 
                    try:
                        pylab.savefig(file_name, bbox_inches=None, pad_inches=0)
                        pylab.close()
                    except:
                        print ('Problem at save_file line 5 -- Skipping this file' + str(j_num))
                        continue   
```
